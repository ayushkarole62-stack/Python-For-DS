# 1 copy()

1.1 deep=True

```python
# deep=True by default ( modificar=tion to the data or indices of the copy
df_copy = df.copy()
```

```python
df.loc["L. Messi", "height_cm"] = 180
```

```python
df
df_copy
```

1.2 deep = false 

```python
# deep = false ( any changes to the data the original will be reflected )
df_shallow_copy = df.copy(deep=False)
```

```python
#updated value in original dataframe 
df.loc["Cristiano Ronaldo", "height_cm"] = 200
```

```python
# shallow copy vs original dataframe 
df
df_shallow_copy
```
