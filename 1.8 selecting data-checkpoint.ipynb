# selecting data by method loc() and iloc()

1. loc() = loc() is label-based, so we have to specify rows and columns based on  their index/row and column labels.


2. iloc() is interger position-based, so we have to specify rows and columns by their integer position value ( 0-based integer)

# 1 Selecting with a single value 

loc[row_label, column_label]

```python
# get all data about L. Messi
df.loc["L. Messi"]
```

```python
# get the height of L.messi
df.loc["L. Messi", "height_cm"]
```

```python
# get the weigth of ronaldo
df.loc["Cristiano Ronaldo", "weight_kg"]
```

# 3 selecting a range of data with a slice

start:stop:step(Note that contrary to usual python slice, both the start and the stop are inculde)

```python
# slice column labels 
players = ["L. Messi", "Cristiano Ronaldo"]

df.loc[players, "age":"club"]
 
```

```python
# slice index labels
column = ["age", "dob", "height_cm", "weight_kg"]

# get top1 and top10 player name 
df.index[:10]

df.loc["L. Messi":"M. Salah", column]

```

# Selecting with conditions

```python
# one comdition: select player with height above 180cm
column = ["age", "dob", "height_cm", "weight_kg"]

df["height_cm"]>180

df.loc[df["height_cm"]>180,column]
```

```python
# multiple condition : select player with height above 180cm from Argentina 


(df["height_cm"]>180) & (df["nationality"]=="Argentina")

df.loc[(df["height_cm"]>180) & (df["nationality"]=="Argentina"), :
]
```

# 4. Selecting with conditions 

iloc cannot accept a boolean series but only a boolean list we have to use the list() function to convert a series into a boolean list 

```python
# one conditiion : select player with heigth above 180cm
# columns =  ["age	dob	height_cm	weight_kg]
columns =  [1,2,3,4]


df.iloc[list(df["height_cm"]>180), columns
]
```

```python
# multiple condition : select player with height above 180cm

df.iloc[list((df["height_cm"]>180) & (df["nationality"]=="Argentina")), :]
```
