# 2. Lambda function + apply() method

```python
# basic function 
def sum_value(a,b):
    x = a+b
    return x
```

```python
sum_value(2,5)
```

```python
# lambda funtion ( one liner)
sum_value_lambda = lambda a,b:a+b
```

```python
sum_value_lambda
```

```python
sum_value_lambda(2,3)
```
